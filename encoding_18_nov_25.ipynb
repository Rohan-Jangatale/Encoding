```python
from sklearn.preprocessing import LabelEncoder
```

```python
import pandas as pd
df=pd.DataFrame({'Color':['red','blue','green','blue','red'],'Target':[1,0,1,0,1]})
df
```

```python
le=LabelEncoder()
df['Color']=le.fit_transform(df['Color'])
df
```

```python
df_ohe=pd.get_dummies(df["Color"],prefix="color")
df_ohe
```

```python
import pandas as pd
df=pd.DataFrame({'Color':['blue','red','green','blue','green','green','blue'],'Target':[1,0,1,0,1,0,0]})
df
```

```python
te=df.groupby("Color")["Target"].mean()
te
```

```python
df["Color"]=df["Color"].map(te)
df
```
